```python
# 定义函数
def match_value_with_time_sequence(original_data, matched_data, field):
    """
    根据时间序列匹配字段值
    :param original_data: 形参1，原始数据
    :param matched_data: 形参2，匹配数据
    :param field: 形参3，匹配的字段名
    """
    for line_original_data in range(0, len(original_data)):
        for line_matched_data in range(0, len(matched_data)):
            if str(matched_data.loc[line_matched_data, '代码']) == str(original_data.loc[line_original_data, '代码']):
                for column in matched_data.columns.to_list()[5: ]:
                    # 更改条件
                    if (str(original_data.loc[line_original_data, '年']) == column[3: 7]
                        and f"{original_data.loc[line_original_data, '月']:02}" == column[-2: ]):
                        original_data.loc[line_original_data, field] = matched_data.loc[line_matched_data, column]
                    else:
                        pass
            else:
                continue
    return original_data
```

```python
# 匹配降水
precipitation_excel = '降水.xlsx'
precipitation_data = pd.read_excel(os.path.join(path, precipitation_excel), sheet_name=0)

original_data = match_value_with_time_sequence(original_data, precipitation_data, '降水')
original_data.head()
original_data.to_excel(os.path.join(path, original_excel), index=False)
```

```python
# 匹配NDVI
ndvi_excel = 'NDVI.xlsx'
ndvi_data = pd.read_excel(os.path.join(path, ndvi_excel), sheet_name=0)

original_data = match_value_with_time_sequence(original_data, ndvi_data, 'NDVI')
original_data.head()
original_data.to_excel(os.path.join(path, original_excel), index=False)
```
